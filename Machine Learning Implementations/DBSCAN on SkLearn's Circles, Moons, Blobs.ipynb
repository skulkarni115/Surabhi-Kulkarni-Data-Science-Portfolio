```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from scipy import spatial
from statistics import mode
import sklearn.metrics.pairwise
from statistics import mean
import random
from collections import defaultdict
import math
```

```python
#load dataset into numpy array
#Circles dataset
#Read file
db_df = pd.read_csv('circle.csv')
display(db_df.head())
db_dict = db_df.to_dict()
```

```python
db_arr = db_df.to_numpy()
#implement distances and getting neighbors
distances = sklearn.metrics.pairwise.euclidean_distances(db_arr,db_arr)
#Define epsilon and min pts
epsilon = .1
min_pts = 45
total_neighbors = []
for i in range(len(db_df['Xcircle_X1'])):
  neighbors = np.argsort(distances[i])[0:(min_pts+5)]
  counter = 0
  for x in neighbors:
    delete_idx = []
    if distances[i][x] > epsilon:
      delete_idx.append(counter)
    counter +=1
  neighbors = np.delete(neighbors,delete_idx)
  total_neighbors.append(neighbors)

```

```python

all_clusters = []
visited = []
for i in range(1000): #len(db_dict['Xcircle_X1']
  if i not in visited:
    neighbors = total_neighbors[i]
    current_cluster = []
    queue = neighbors
    queue = list(queue)
    if len(neighbors) < min_pts:
      visited.append(i)
      continue
    while queue:
      num = queue.pop(0)
      if num not in visited:
        visited.append(num)
        current_cluster.append(num)
        n_neigh = total_neighbors[num]
        if len(n_neigh)>= min_pts:
          queue.extend(n_neigh)
      else:
        continue
  else:
    continue
  #print(len(current_cluster))
  all_clusters.append(current_cluster)
```

```python
print(len(all_clusters))

total_len = 0
for i in range(len(all_clusters)):
  total_len += len(all_clusters[i])
print(total_len)
print(len(total_neighbors))
print(len(visited))
```

```python
db_df.insert(0, "cluster", 'nan')
```

```python
for i in range(len(all_clusters)):
  for x in all_clusters[i]:
    db_df['cluster'][x] = (i+1)
```

```python
display(db_df.head())
db_df.plot.scatter(x='Xcircle_X1',y='Xcircle_X2', c = 'cluster', colormap = 'viridis')
```
